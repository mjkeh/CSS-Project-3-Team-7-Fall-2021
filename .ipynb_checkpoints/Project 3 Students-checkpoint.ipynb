```python
# Split data into train/validation sets
np.random.seed(10)

X_train, X_validate, y_train, y_validate = train_test_split(X, y, train_size = .80, test_size=0.20,
                                                   stratify=y)
```

### Model 1. Logistic Regression

Logistic regression is a parametric method that uses a logistic function to model a binary outcome. Regression coefficients are estimated using maximum likelihood estimation, and represent the log-odds of the outcome. These coefficients can be converted to probabilities and used to make predictions. The model assumes that the log-odds of the binary outcome is a linear combination of the predictors. 

Pros: simple to implement, interpretable

Cons: assumes linear relationship between outcome and predictors

```python
# logistic regression
logit_reg = LogisticRegression()

#hyperparameter grid
import warnings
from sklearn.exceptions import DataConversionWarning
warnings.filterwarnings(action='ignore')
from sklearn.metrics import accuracy_score

param_grid = {'penalty': ['l1', 'l2', 'elasticnet'],
             'C': np.arange(.1, 1, .1),
               'fit_intercept': ['True', 'False'],
             'solver': ['liblinear', 'saga']}

logit_grid = GridSearchCV(logit_reg, param_grid, cv=3)
logit_grid.fit(X_train, y_train)
""
best_index = np.argmax(logit_grid.cv_results_["mean_test_score"])
best_logit_pred = logit_grid.best_estimator_.predict(X_validate)

print("Optimal hyperparameters for Logistic Regression:", logit_grid.cv_results_["params"][best_index])
```

### Model 2. Support Vector Machine

SVM is a non-parametric method that determines the best separating hyperplane (largest margin) between two classes. 

Pros: flexible; specfically, it allows X's to have higher order polynomials, interactions, or other functions of the original X's. 

Cons: computationally intensive since it relies on higher dimensional space where data are linearly seperable. 

```python
# SVM
svm = SVC()

#hyperparameter grid
param_grid = {'C': [.5,1],
              'gamma': [1,.1]}

svm_grid = GridSearchCV(svm, param_grid, cv=3)
svm_grid.fit(X_train, y_train)

best_index = np.argmax(svm_grid.cv_results_["mean_test_score"])
best_svm_pred = svm_grid.best_estimator_.predict(X_validate)

print("Optimal hyperparameters for SVM:", svm_grid.cv_results_["params"][best_index])
```

### Model 3. Random Forest

Random Forest is an extension of the decision tree method. A random forest fits a number of decision tree classifiers on various sub-samples of the dataset and uses averaging to improve the predictive accuracy and control over-fitting (the "bagging" method). A subset of features is also randomly selected at each split. 

Pros: deals well with non-linearities and complex configurations of features, relatively interpretable, and can  be used to estimate variable importance. 

Cons: greedy algorithm, can result in high variance/overfitting.

```python
# Random Forest
rf_classifier = RandomForestClassifier()

#hyperparameter grid
param_grid = {
    'criterion': ['gini', 'entropy'],
    'bootstrap': ['True', 'False'],
    'min_samples_leaf': [1,2,3],
    'min_samples_split': [2,3,4],
}

rf_grid = GridSearchCV(estimator = rf_classifier, param_grid = param_grid, cv = 3)
rf_grid.fit(X_train, y_train)

best_index = np.argmax(rf_grid.cv_results_["mean_test_score"])
best_rf_pred = rf_grid.best_estimator_.predict(X_validate)

print("Optimal hyperparameters for Random Forest:", rf_grid.cv_results_["params"][best_index])
```

## 3.3 - Validation Metrics

**Hint**: Try writing a for loop to use [`cross_val_score()`](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html) to check for accuracy, precision, recall and f1 across all of your models.

```python
for clf, label in zip([logit_grid.best_estimator_, 
                       svm_grid.best_estimator_, 
                       rf_grid.best_estimator_], ['Logistic Regression', 
                                            'SVM', 
                                            'Random Forest']):
    scores = cross_val_score(clf, X_validate, y_validate.ravel(), scoring='accuracy', cv=5)
    print('Accuracy: %0.2f [%s]' % (scores.mean(), label))
    scores = cross_val_score(clf, X_validate, y_validate.ravel(), scoring='precision', cv=5)
    print('Precision: %0.2f [%s]' % (scores.mean(), label))
    scores = cross_val_score(clf, X_validate, y_validate.ravel(), scoring='recall', cv=5)
    print('Recall: %0.2f [%s]' % (scores.mean(), label))
    scores = cross_val_score(clf, X_validate, y_validate.ravel(), scoring='f1', cv=5)
    print('F1: %0.2f [%s]' % (scores.mean(), label))       
```

From the scores above, we are primarily prioritizing Recall score. There is a tie between the Logistic Regression and Random Forest models for their Recall scores. We ultimately selected the Logistic Regression Model as our best model after weighing the importance of identifying true positives.

```python
# Predict the outcomes and the probabilities of the validation set

best_model_pred = hyper_logit_model.predict(X)
best_model_prob = hyper_logit_model.predict_proba(X)

d= {"Probabilities":list(best_model_prob),"Prediction":list(best_model_pred)}
df = pd.DataFrame(d)
df['prob_pass']='Nan'
df['prob_fail']='Nan'

for i in range(0,len(df)):
    df['prob_pass'][i]=df['Probabilities'][i][0]
    df['prob_fail'][i]=df['Probabilities'][i][1] 
    
df.head()
```

```python
# Order audits by their probability of detecting a "Fail" score

df.sort_values(by='prob_fail', axis=0, ascending=False, inplace=True, kind='quicksort', na_position='last')
df['prob_fail'] = df['prob_fail'].astype("float64")
df['prob_pass'] = df['prob_pass'].astype("float64")
```

```python
# Plot distribution of pass/fail among the first 1000 observations in the dataset
df_1000 = df.nlargest(1000,['prob_fail']) 

sns.displot(df_1000, x="prob_fail", bins=10)

plt.title("Probability of failing for the top 1000 cases for failing")
plt.show()

```

```python
# Simulate random audits on the full chicago_2011_to_2013.csv dataset by picking 1000 observations at random
X = pd.DataFrame(X)
X_train = pd.DataFrame(X_train)
X_validate = pd.DataFrame(X_validate)
y_train = pd.DataFrame(y_train)
y_validate = pd.DataFrame(y_validate)


sample_1000 =X.sample(1000)
sample_1000_pred = hyper_logit_model.predict(sample_1000)
sample_1000_prob = hyper_logit_model.predict_proba(sample_1000)

S= {"Probabilities":list(sample_1000_prob),"Prediction":list(sample_1000_pred)}
df_random = pd.DataFrame(S)
df_random['prob_pass']='Nan'
df_random['prob_fail']='Nan'

for i in range(0,len(df_random)):
    df_random['prob_pass'][i]=df_random['Probabilities'][i][0]
    df_random['prob_fail'][i]=df_random['Probabilities'][i][1] 
    
df_random.head(10)
```

```python
sns.displot(df_random, x="prob_fail", bins=10)
plt.title("Probability of failing for random 1000 cases")
plt.show()

```

Comparing rescults, the audit of 1,000 random cases proved to be a little skewed. This matched the the general trend of the data set.
